```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(loc = 0, scale = 1, size = 1000)
# 랜덤한 정규화된 데이터 1000개 생성 
# loc= 균등분포에서 제일 작은값 , scale = 균등분포 제일 끝값끼리의 차) 

# data 분포 시각화
sns.histplot(data)
plt.show()

# data 통계 지표 계산
mean = np.mean(data)
std = np.std(data)
print("mean:", mean)
print("std:", std)
```

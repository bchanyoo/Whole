# 인공신경망 가중치를 적용한 시각화 실습

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2차원 가중치 행렬을 랜덤으로 초기화
w = torch.randn(3,3)
print(w)

# flatten을 이용해 1차원 벡터로 변환
w_flattened = w.flatten()
print(w_flattened)

#시각화
plt.hist(w_flattened, bins = 20)
plt.xlabel('Weight Value')
plt.ylabel('Frequendcy')
plt.show()
```

## 크기가 1인 랜덤한 값을 가지는 bias를 생성




```python
b = torch.rand(1)
print(b)
#시각화
plt.hist(b.detach().numpy(), bins = 20) #b.detach().numpy() => tensor -> numpy 배열로 변환하는 코드
plt.xlabel('Bias Value')
plt.ylabel('Frequency')
plt.show()
```

# Mean Squared Error 간단 실습

```python
def mean_squared_error(y_true, y_pred):
  mean_temp = np.mean((y_true - y_pred) ** 2)
  return mean_temp

y_true = np.array([1,2,3,4])
y_pred = np.array([2,3,5,6])

mean_value = mean_squared_error(y_true, y_pred)
print(mean_value)
```

## binary cross entropy 간단한 실습

```python
def binary_cross_entropy_loss(y_true_temp, y_pred_temp):
  epsilon = 1e-5
  n = y_true_temp.shape[0]
  print('n value>>', n)
  loss = -np.mean(y_true_temp * np.log(y_pred_temp + epsilon) + (1 - y_true_temp) * np.log(1 - y_pred_temp + epsilon))
  return loss

y_true_temp = np.array([1,0,1,1])
y_pred_temp = np.array([0.99, 0.1, 0.8, 0.7])
loss = binary_cross_entropy_loss(y_true_temp, y_pred_temp)
print(loss)
```

# Cross Entropy Loss 간단한 실습

```python
def cross_entropy_loss(y_true_01, y_pred_01, c): #c는 클래스의 갯수
    '''
    y_true_01 -> 실제값(크기 : n, c)
    y_pred_01 -> 예측값(크기 : n, c)
    c -> 클래스 갯수
    '''
    n_temp = y_true_01.shape[0]
    print('n value >>>', n_temp)
    ce_loss = -1 / n_temp * np.sum(y_true_01 * np.log(y_pred_01))

    return ce_loss

y_true_01 = np.array([[1, 0, 0],[0, 1, 0], [0, 0 ,1]])
y_pred_01 = np.array(([0.72, 0.2, 0.1], [0.1, 0.8, 0.1], [0.2, 0.3, 0.5]))
c = 3

ce_loss_value = cross_entropy_loss(y_pred_01, y_pred_01, c)
print(ce_loss_value)
```

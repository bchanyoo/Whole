```python
#작은 그림판 만들기
plt.subplot(2,1,1)     # rows, columns, panel number
plt.plot(x, np.sin(x))
plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, np.sin(x), color = 'b') #blue의 약어
plt.plot(x, np.sin(x-1), color = 'g') #green의 약어어
plt.plot(x, np.sin(x-2), color = '0.75') #1은 검정색, 채도를 표현.
plt.plot(x, np.sin(x-3), color = '#FFDD44') #FF/DD/44 -> 노란색
plt.plot(x, np.sin(x-4), color = (1.0, 0.2, 0.3)) # 빨간색. RGB 중 R이 1.0, 나머지는 0.2, 0.3
plt.plot(x, np.sin(x-5), color = 'chartreuse') #거의 사용하지 않음. 연두색의 색상명
```

```python
#line의 종류
plt.plot(x, x+0, linestyle = 'solid') 
plt.plot(x, x+1, linestyle = 'dashed') #_ _ _ _ _
plt.plot(x, x+2, linestyle = 'dashdot') # -.-.-.
plt.plot(x, x+3, linestyle = 'dotted') # . . . . . . .

#기호
plt.plot(x, x+4, linestyle = '-') 
plt.plot(x, x+5, linestyle = '--') #_ _ _ _ _
plt.plot(x, x+6, linestyle = '-.') # -.-.-.
plt.plot(x, x+7, linestyle = ':') # . . . . . . .
```

```python
plt.plot(x, x+0, '-g')    #solid green
plt.plot(x, x+1, '--c')   #dashed cyan
plt.plot(x, x+2, '-.k')   #dashdot black
plt.plot(x, x+3, ':r')    #dotted red
```

```python
#세부 조정하는 법
plt.plot(x, np.sin(x)) #x축의 한 칸이 2, y축의 한 칸이 0.5임(기준이 다름)
#기준을 설정하지 않으면 최대한 꽉 채워서 보려주려 한다. 실제 파형과 다름.
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x))
plt.xlim(10,0) #좌우가 반전되어 나옴.
plt.ylim(1.2, -1.2) #상하반전
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5])
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight') #그래프 타이트하게 출력, 기본값임. but 실제의 파형과는 다름.
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')   #x,y축의 비율을 동일하게. 실제 파형을 가늠하려면 equal 함수 사용.
```

```python
# label 붙이기
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')   #title 주기
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x), '-g', label = 'sin(x)') #선에 labelling한 것.
plt.plot(x, np.cos(x), ':b', label = 'cos(x)')
plt.axis('equal')
plt.legend() #그래프에 대한 설명
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine')
```

```python
#산점도의 표현, 점을 찍어 데이터가 얼마나 분산되어있는지 보여주는 것 (scatter plot)
x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x, y, 'o', color = 'k') #marker, o는 점
```

```python
#marker 사용법, 종류
plt.style.use('classic') #style 지정
rng = np.random.RandomState(0) #넘파이에서의 랜덤 시드값 = (0)
for marker in ['o', ',', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']: #o마커, ,마커 < 엄청 작음. x, +마커. 4방향 삼각형 기호, square, diamond
  plt.plot(rng.rand(5), rng.rand(5), marker, label = 'marker={0}'.format(marker)) #마커 라벨링
  plt.legend(numpoints = 1) #우상단의 마커 설명을 하나만 출력해주라는 뜻. 기본값은 2개씩 나옴.
  plt.xlim(0,1.8) #x축을 1.8까지 늘리는 것.
```

```python
#산점도의 선 연결
plt.plot(x, y, '-o') #-o marker사용, --o도 사용 가능. --ok도 가능.k는 블랙.
```

```python
#marker 편집하기 + p 마크(pentagon)
plt.plot(x, y, '-p', color = 'gray',                 #-선, pentagon 마크
         markerfacecolor = 'white',                  #마크색
         linewidth = 4,                              #굵기
         markersize = 15,                            #마크 크기
         markeredgecolor = 'red',                    #마크 테두리
         markeredgewidth = 2)                        #마크 테두리 굵기     
```

```python
#rand = 0~1의 값을 지정된 숫자만큼 임의의 값을 만들어 낸다.
#randn = 0을 기준으로 편차만큼 떨어진 임의값ㅇㄹ 지정된 숫자만큼 만들어 냄. +-
#randint = 지정된 범위 안의 정수 1개를 만들어 낸다. 
#점의 크기 키우기
plt.style.use('classic')
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
color = rng.rand(100) #100개의 컬러를 랜덤으로 만듦.
sizes = rng.rand(100) * 1000 #사이즈는 -값이 없다. 블랙홀 제외. 0~1 사이의 수라서 작기때문에 곱하기 천을 해줌.
plt.scatter(x, y, s = sizes, c = color, alpha = 0.3, cmap = 'viridis') #알파 명령어. 70프로 투명해짐, cmap은 컬러맵. 색상의 일괄성+크기에 따라 매겨짐.
```

```python
#scikit-learn
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
#features #피쳐가 네개 있음(iris 길이,폭 / iris길이, 폭)
plt.scatter(features[0], features[1], c = iris.target, s = features[3]*100, alpha = 0.2, cmap = 'viridis') #가로, 세로
plt.xlabel(iris.feature_names[0]) #데이터 이름찾기 = iris.feature_names[0](꽃받침 길이)
plt.ylabel(iris.feature_names[1])
iris.target #scikit-learn에서는 타겟이 라벨임, 라벨 종류 0,1,2 (즉 세 종류가 있다)
iris.feature_names[3]
```

```python
#오차의 시각화
x = np.linspace(0,10,50) #0에서 10까지 균등하게 50개로 나눔.
dy = 0.8 #dy = error의 range 오차범위는 0.8
y = np.sin(x + np.random.randn(50))
plt.errorbar(x, y, yerr=dy, fmt = '.k') #오차범위를 나타내는 errorbar, fmt는 포맷.yerr은 세로로 표현돼야하니 y축에 대한 에러값 
```

```python
plt.errorbar(x, y, yerr=dy, fmt = 'o', c = 'k', ecolor = 'lightgray', elinewidth = 3) #ecolor = 에러바 컬러, elinewidth - 에러바 
```
